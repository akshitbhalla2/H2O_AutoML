# Importing Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading and Exploring the Data

```python
data = pd.ExcelFile("bank_term_deposit_marketing_analysis.xlsx")
```

```python
data.sheet_names
```

```python
client_info =  pd.read_excel("bank_term_deposit_marketing_analysis.xlsx", sheet_name = "CLIENT_INFO")
loan_history =  pd.read_excel("bank_term_deposit_marketing_analysis.xlsx", sheet_name = "LOAN_HISTORY")
marketing_history =  pd.read_excel("bank_term_deposit_marketing_analysis.xlsx", sheet_name = "MARKETING HISTORY")
subscription_history =  pd.read_excel("bank_term_deposit_marketing_analysis.xlsx", sheet_name = "SUBSCRIPTION HISTORY")
```

```python
df = pd.merge(client_info, loan_history, on = ["ID"])
df = pd.merge(df, marketing_history, on = ["ID"])
df = pd.merge(df, subscription_history, on = ["ID"])
```

```python
df
```

```python
df.info()
```

```python
df = df.drop(columns = ["ID"])
```

# Data Prep & Start H2O

```python
import h2o
h2o.init()
```

```python
df_h2o = h2o.H2OFrame(df)
```

```python
train, test = df_h2o.split_frame(ratios = [0.75])
x = train.columns
y = "TERM_DEPOSIT"
x.remove(y)
```

# Run H2O AutoML

```python
from h2o.automl import H2OAutoML
```

```python
df["TERM_DEPOSIT"].value_counts() # Class imbalance
```

```python
aml = H2OAutoML(
    max_runtime_secs = 900,
    balance_classes = True,
    stopping_metric = "logloss",
    project_name = "Final",
    seed = 1
)
```

```python
aml.train(
    x = x,
    y = y,
    training_frame = train
)
```

# AutoML Leaderboard and Ensemble Exploration

```python
lb = aml.leaderboard
lb.head(rows = lb.nrows)
```

```python
se = aml.leader
metalearner = h2o.get_model(se.metalearner()["name"])
```

```python
metalearner.varimp()
```
